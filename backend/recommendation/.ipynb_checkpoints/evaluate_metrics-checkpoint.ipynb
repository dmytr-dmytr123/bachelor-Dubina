```python
import joblib
import json
import numpy as np
from recommender import recommend_events
from vectorizer1 import EventVectorizer
from lightgbm import LGBMRegressor
```

```python
with open("test_input_full_features.json", "r", encoding="utf-8") as f:
    test_samples = json.load(f)

vectorizer, model = joblib.load("model.pkl")
print(f"Loaded {len(test_samples)} samples")
```

```python
def precision_at_k(predicted_ids, relevant_ids, k):
    predicted_k = predicted_ids[:k]
    relevant_set = set(relevant_ids)
    hits = sum([1 for pid in predicted_k if pid in relevant_set])
    return hits / k

def recall_at_k(predicted_ids, relevant_ids, k):
    predicted_k = predicted_ids[:k]
    relevant_set = set(relevant_ids)
    hits = sum([1 for pid in predicted_k if pid in relevant_set])
    return hits / len(relevant_ids) if relevant_ids else 0.0
```

```python
def evaluate_precision_recall_k(model, vectorizer, test_samples, k=5):
    precisions, recalls = [], []

    for sample in test_samples:
        user_profile = sample["user_profile"]
        user_history = sample.get("user_history", [])
        candidates = sample["candidates"]
        ground_truth_ids = sample["ground_truth_ids"]

        recs = recommend_events(
            user_profile=user_profile,
            events=candidates,
            user_history=user_history,
            vectorizer=vectorizer,
            model=model,
            top_n=k
        )

        predicted_ids = [str(rec["event"].get("_id")) for rec in recs]

        precisions.append(precision_at_k(predicted_ids, ground_truth_ids, k))
        recalls.append(recall_at_k(predicted_ids, ground_truth_ids, k))

    avg_precision = np.mean(precisions)
    avg_recall = np.mean(recalls)

    print(f"Precision@{k}: {avg_precision:.4f}")
    print(f"Recall@{k}: {avg_recall:.4f}")
    return avg_precision, avg_recall

# Run it
evaluate_precision_recall_k(model, vectorizer, test_samples, k=5)
```

```python
import matplotlib.pyplot as plt

ks = [1, 3, 5, 10]
precisions, recalls = [], []

for k in ks:
    p, r = evaluate_precision_recall_k(model, vectorizer, test_samples, k=k)
    precisions.append(p)
    recalls.append(r)

plt.plot(ks, precisions, label="Precision@K")
plt.plot(ks, recalls, label="Recall@K")
plt.xlabel("K")
plt.ylabel("Score")
plt.title("Precision and Recall at K")
plt.legend()
plt.grid(True)
plt.show()
```
