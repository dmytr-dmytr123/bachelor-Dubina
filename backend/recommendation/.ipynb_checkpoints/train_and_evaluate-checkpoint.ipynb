```python
import json
from lightgbm import LGBMRegressor
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

from vectorizer1 import EventVectorizer
from semantic_helper import SemanticEmbeddingHelper
from time_decay import get_time_weight
```

```python
with open("interactions_scored.json", "r", encoding="utf-8") as f:
    interactions = json.load(f)

print(f"Loaded {len(interactions)} interactions")
```

```python
vectorizer1 = EventVectorizer()
vectorizer1.fit(interactions)
entry = interactions[0]

print(len(vectorizer1.transform_user_profile(entry['user_profile'])))
print(len(vectorizer1.transform_user_history(entry['user_history'])))
print(len(vectorizer1.transform_event(entry['event'])))


X, y = [], []

for entry in interactions:
    user_vec = vectorizer1.transform_user_profile(entry["user_profile"])
    history_vec = vectorizer1.transform_user_history(entry["user_history"])
    event_vec = vectorizer1.transform_event(entry["event"])

    print(
        f"user_vec: {np.shape(user_vec)}, "
        f"history_vec: {np.shape(history_vec)}, "
        f"event_vec: {np.shape(event_vec)}"
    )

    features = np.concatenate([user_vec, history_vec, event_vec]).flatten()
    expected_length = vectorizer1.user_vector_size() + vectorizer1.history_vector_size() + vectorizer1.event_vector_size()
    
    if len(features) != expected_length:
        print(f"❌ Skipping due to unexpected feature length: {len(features)} (expected: {expected_length})")
        continue
    
    X.append(features)
    y.append(entry["score"])



X = np.array(X)
y = np.array(y)

print(f"✅ Feature vector shape: {X.shape}")
```

```python
import matplotlib.pyplot as plt

plt.hist(y, bins=20)
plt.title("Розподіл цільової змінної (y)")
plt.xlabel("score")
plt.ylabel("кількість прикладів")
plt.show()
```

```python
threshold = 0.05
X_filtered = []
y_filtered = []

for xi, yi in zip(X, y):
    if yi > threshold:
        X_filtered.append(xi)
        y_filtered.append(yi)

X = np.array(X_filtered)
y = np.array(y_filtered)

print(f"📉 Залишилось прикладів: {len(y)}")
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from lightgbm import LGBMClassifier
import matplotlib.pyplot as plt
import seaborn as sns

# Бінаризуємо цільову змінну ДО train_test_split
y_binary = (y > 0.3).astype(int)

# Потім розбиваємо
X_train, X_val, y_train, y_val = train_test_split(X, y_binary, test_size=0.2, random_state=42)

# Тренуємо класифікатор
model = LGBMClassifier()
model.fit(X_train, y_train)

# Прогноз і метрики
y_pred = model.predict(X_val)

print("✅ Accuracy:", accuracy_score(y_val, y_pred))
print("\n📊 Classification report:\n", classification_report(y_val, y_pred))

```

```python
cm = confusion_matrix(y_val, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
